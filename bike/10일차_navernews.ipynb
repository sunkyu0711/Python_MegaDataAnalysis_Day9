```python
# 워드클라우드로 시각화

from konlpy.tag import Okt
import collections

url='https://news.naver.com/breakingnews/section/101/259'
res=requests.get(url).text
soup=BeautifulSoup(res)
temp=soup.select_one('ul.sa_list').select('li',limit=5)

news_list=[]
for li in temp:
    news_info={'title':li.select_one('strong.sa_text_strong').text,
              'date':li.select_one('div.sa_text_datetime.is_recent').text,
              'news_url':li.select_one('a')['href']}
    news_list.append(news_info)

for news in news_list:
    news_url=news['news_url']
    # print(news_url)
    res=requests.get(news_url).text
    soup=BeautifulSoup(res)
    body=soup.select_one('article.go_trans._article_content') # 주의: select 함수를 이용할 때에는 공백을 없애야(.으로 메우면 됨)
    news_content=body.text.replace('\n','').strip()
    news['news_content']=news_content

df=pd.DataFrame(news_list)

okt=Okt()
clist=[]
for word in df['news_content']:
    token=okt.pos(word)
    for word, tag in token:
        if tag in ['Noun','Adjective']:
            clist.append(word)
counts=collections.Counter(clist)
tag=counts.most_common(50)
print(tag)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

url='https://news.naver.com/breakingnews/section/101/259'
res=requests.get(url).text
soup=BeautifulSoup(res)
temp=soup.select_one('ul.sa_list').select('li',limit=5)

news_list=[]
for li in temp:
    news_info={'title':li.select_one('strong.sa_text_strong').text,
              'date':li.select_one('div.sa_text_datetime.is_recent').text,
              'news_url':li.select_one('a')['href']}
    news_list.append(news_info)

for news in news_list:
    news_url=news['news_url']
    # print(news_url)
    res=requests.get(news_url).text
    soup=BeautifulSoup(res)
    body=soup.select_one('article.go_trans._article_content') # 주의: select 함수를 이용할 때에는 공백을 없애야(.으로 메우면 됨)
    news_content=body.text.replace('\n','').strip()
    news['news_content']=news_content

df=pd.DataFrame(news_list)

okt=Okt()
clist=[]
for word in df['news_content']:
    token=okt.pos(word)
    for word, tag in token:
        if tag in ['Noun','Adjective']:
            clist.append(word)
counts=collections.Counter(clist)
tag=counts.most_common(50)

fpath='C:\Windows\Fonts\malgunbd.ttf'
wc=WordCloud(font_path=fpath,background_color='white') # width, height 등의 설정을 해도 됨
cloud=wc.generate_from_frequencies(dict(tag))

plt.figure(figsize=(10,8))
plt.imshow(cloud)
```

```python
from wordcloud import STOPWORDS

s_words=STOPWORDS.union({'이','비','내','날','전','위','의','현','원','체','수','사','은','것','를','제','로','앞','고','율'})

url='https://news.naver.com/breakingnews/section/101/259'
res=requests.get(url).text
soup=BeautifulSoup(res)
temp=soup.select_one('ul.sa_list').select('li',limit=5)

news_list=[]
for li in temp:
    news_info={'title':li.select_one('strong.sa_text_strong').text,
              'date':li.select_one('div.sa_text_datetime.is_recent').text,
              'news_url':li.select_one('a')['href']}
    news_list.append(news_info)

for news in news_list:
    news_url=news['news_url']
    # print(news_url)
    res=requests.get(news_url).text
    soup=BeautifulSoup(res)
    body=soup.select_one('article.go_trans._article_content') # 주의: select 함수를 이용할 때에는 공백을 없애야(.으로 메우면 됨)
    news_content=body.text.replace('\n','').strip()
    news['news_content']=news_content

df=pd.DataFrame(news_list)

okt=Okt()
clist=[]
for word in df['news_content']:
    token=okt.pos(word)
    for word, tag in token:
        if tag in ['Noun','Adjective']:
            clist.append(word)
counts=collections.Counter(clist)
tag=counts.most_common(100)

fpath='C:\Windows\Fonts\malgunbd.ttf'
wc=WordCloud(font_path=fpath,background_color='white',stopwords=s_words) # 필요 없는 단어 없애기... 안 빠졌네?
cloud=wc.generate_from_frequencies(dict(tag))

plt.figure(figsize=(10,8))
plt.axis('off')
plt.imshow(cloud)
```
